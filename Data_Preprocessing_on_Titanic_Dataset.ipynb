# **Checking and Removing Nulls**

```python
df.isna().sum()
```

**We don't require Cabin data as it doesn't affect the model and also it has so many null values as we remove rows there will be a huge data loss.**

```python
df.drop(columns={"Cabin"}, inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

**We can't remove age column as it is essential feature in modeling. So we need to impute the age column by its mean.**

```python
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
```

```python
df
```

```python
df.isna().sum()
```

**Now we have categorical data so we need to fill the nulls to mode of the column**

```python
mode_embark = df['Embarked'].mode()[0]
df['Embarked'] =df['Embarked'].fillna(mode_embark)
df
```

```python
df.isna().sum()
```

# **Now the Nulls are gone we need to change the categorical data to numerical data**

```python
df.dtypes
```

so there are 4 columns that are categorical - Name, Sex, Ticket, Embarked. In that Name and ticket are baseless feature (no use) and also PassengerId (which is int but we dont need that)

```python
df.drop(columns={"Name","PassengerId","Ticket"},inplace=True)
```

So we are remained with Sex and Embarked and both are nominal only we need to use One Hot Encoding

```python
pd.get_dummies(df,drop_first=True)
```

# **By this Every Feature is numerical**

if we have any ordinal feature in data and we need to label encode then we need to use  ---  pd.DataFrame.map(mapping_dict)

```python
df = pd.DataFrame({
  'T-shirt_Size': ['S', 'S', 'M', 'L', 'XL', 'XXL'],
  'Price': [300, 400, 350, 400, 400, 450],
  'Brand': ['Levis', 'GAP', 'Levis', 'Levis', 'Levis', 'GAP'],
  'Label': [0, 0, 1, 1, 0, 0]})

df
```

```python
mapping_dict= {
    "S" : 1,
    "M": 2,
    "L" : 3,
    "XL" : 4,
    "XXL" : 5
}

df['T-shirt_Size'] = df['T-shirt_Size'].map(mapping_dict)
```
